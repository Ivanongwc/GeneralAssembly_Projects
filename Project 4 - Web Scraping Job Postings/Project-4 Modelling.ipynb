### Perform Data Cleaning

```python
# Dropping unwanted/unecessary column

df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Create filters for data-related jobs

job_title_filter = ["data", "scientist", "science", "analyst", "analytics", "business", "financial", "visualisation", "visualization", "engineer", "machine learning", "deep learning", "big data", "intelligence"]
```

```python
# Filtering job titles which have key words from the above list
keep=[]

for row in range(len(df)):
    for i in job_title_filter:
        if i in df['Job_title'][row].lower():
            if row not in keep:
                keep.append(row)

df = df.iloc[keep]
df = df.reset_index(drop=True)
df.shape
```

```python
# Handling expiry date and job posted date

df['Expiry_date']=df['Expiry_date'].str.replace('Closing on','')
df['Last_posted_date']=df['Last_posted_date'].str.replace('Posted','')

df['Expiry_date'] = pd.to_datetime(df['Expiry_date'])
df['Last_posted_date'] = pd.to_datetime(df['Last_posted_date'])
```

### Exploring the target variable: Salary_range

```python
# Cleaning salary columns
df['Salary_range'] = df['Salary_range'].str.replace('$','').str.replace(',','')

# New column for min salary and max salary
df['Salary_min'] = [i.split('to')[0] for i in df['Salary_range']]
df['Salary_max'] = [i.split('to')[-1] for i in df['Salary_range']]

# Dropping Salary_range col as redundant
df.drop('Salary_range',inplace=True,axis=1)
```

```python
df[['Salary_min','Salary_max']].head(5)
```

```python
# Setting salary type
df['Salary_min'] = df['Salary_min'].astype(int)
df['Salary_max'] = df['Salary_max'].astype(int)
```

```python
# Creating a column for numeric monthly / annual salary
df['Salary_frequency'] = df['Salary_frequency'].apply(lambda x: 1 if x=='Monthly' else 12 if x=='Annually' else 0)

salary_col = ['Salary_min','Salary_max']

# Converting salary columns to monthly salary
for col in salary_col:
    df[col] = df[col] / df['Salary_frequency']
```

### Staring by setting the salary to be a Classification Problem

```python
# Starting with setting target as a classification problem
# For Average salaries equal and above the median of 6000, it is classified as high
# For average salaries below 6000, it is classified as low.

def f(row):
    if row['Salary_avg'] >= 6000:
        val = 1
    else:
        val = 0
    return val
```

```python
# Create a new column for salary class:
df['Salary_class'] = df.apply(f, axis=1)
```

```python
# Comparing the target size imbalance:
df['Salary_class'].value_counts()
```

```python
# Baseline - How likely is it to predict the majority class value correctly?
print('Baseline score = {}'.format(682/1333))
```

To build simple models first and subsequently include other features to see if predictability of salary increases

### By Job Title

```python
# Job title
df['Job_title'].value_counts().head(20)
```

```python
# Creating a feature set based on the Title to predict salary

cvec = CountVectorizer(min_df=10,max_df=1200)
job_titles = cvec.fit_transform(df['Job_title'])
```

```python
job_titles.shape
```

```python
# Setting it into a data frame

job_titles = pd.DataFrame(job_titles.todense(), columns=cvec.get_feature_names())


# Performing train, test, split

y = df['Salary_class']
X = job_titles

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

__Using Logistic Regression__

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

cv_score = cross_val_score(logreg, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1)
print(cv_score)
print('CV Score :', cv_score.mean())

# Scoring with the test set

from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, logreg.predict_proba(X_test)[:,1])
print('ROC AUC Score: {:.5f}'.format(auc))
print('Accuracy of logistic regression classifier on test set: {:.5f}'.format(logreg.score(X_test, y_test)))
```

```python
# Optimising via GridSearch 
C_vals = [.0001,.001,.01,.05, .1,.12,.13,.135,.15,.16,.17,1,3,5,10,20,50,100]
penalties = ['l1','l2']

gs = GridSearchCV(logreg, {'penalty': penalties, 'C': C_vals}, verbose=False, cv=5)
gs.fit(X_train, y_train)

y_pred = logreg.predict_proba(X_test)
y_pred_lr1 = gs.best_estimator_.predict(X_test)
y_predp_lr1 = gs.best_estimator_.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred_lr1, target_names=["Low Salary","High Salary"]))
print('Accuracy of logistic regression classifier on test set: {:.5f}'.format(logreg.score(X_test, y_test)))
```

```python
acc_lr1 = accuracy_score(y_pred_lr1, y_test)
print("Accuracy after GridSearch:", acc_lr1)
roc_lr1 = roc_auc_score(y_test, y_predp_lr1)
print("ROC_AUC after GridSearch:", roc_lr1)
```

__Using Random Forest Regression__

```python
from sklearn.ensemble import RandomForestClassifier
rfreg = RandomForestClassifier(n_estimators=100, random_state=42)

scores = cross_val_score(rfreg, X_train, y_train, cv=10)

print("Cross-validated scores:", scores)
print("Mean Cross-validated scores:", np.mean(scores))
print("Std Cross-validated scores:", np.std(scores))

rfreg.fit(X_train, y_train)

print("Train score:", rfreg.score(X_train, y_train))
print("Test score:", rfreg.score(X_test, y_test))
print("Train Mean Squared Error:", mean_squared_error(y_train, rfreg.predict(X_train)))
print("Train Mean Squared Error:", mean_squared_error(y_test, rfreg.predict(X_test)))
```

```python
y_pred_rf1 = rfreg.predict(X_test)
y_predp_rf1 = rfreg.predict_proba(X_test)[:,1]
print(classification_report(y_test, y_pred_rf1, target_names=["Low Salary","High Salary"]))
acc_rf1 = accuracy_score(y_pred_rf1, y_test)

# ROC AUC Scoring
roc_rf1 = roc_auc_score(y_test, y_predp_rf1)
print("ROC AUC Score:", roc_rf1)
print('Accuracy of logistic regression classifier on test set: {:.5f}'.format(accuracy_score(y_pred_rf1, y_test)))
```
